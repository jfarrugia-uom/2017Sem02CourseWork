```python
from bs4 import BeautifulSoup
import urllib
r = urllib.urlopen('https://en.wikipedia.org/wiki/Big_data').read()
soup = BeautifulSoup(r, "html5lib")
print type(soup)
```

```python
print soup.prettify()[0:10000]
```

```python
all_links = soup.find_all('a')
all_links
```

```python
for link in all_links:
    print ("%s\t%s" % (link.get("href"), link.get("class")))
```

```python
soup.find_all(id="t-info")[0].a["href"]
```

```python
# find number of sections
len(soup.find_all(class_="mw-headline"))
```

```python
# find article title
print(soup.find_all(class_="firstHeading")[0].text)
```

```python
text = soup.find(id="mw-content-text").get_text(" ", strip=True)
words = [word.lower() for word in text.split()]
len(words)
```
